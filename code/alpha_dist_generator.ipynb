### Iterating through train-test split random seeds to generating optimal `RidgeCV` alphas

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# import data
X_train = pd.read_csv('datasets/X_train.csv')
X_test = pd.read_csv('datasets/X_test.csv', )
y_train = pd.read_csv('datasets/y_train.csv')

# scale
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
Z_train = ss.fit_transform(X_train)
Z_test = ss.transform(X_test)

# t/t split random_seed over range(1000), fit RidgeCV and append .alpha_ to [alphas]
alphas = []
for tsplit in range(500):
    Zs_train, Zs_test, ys_train, ys_test = train_test_split(Z_train, y_train, random_state=tsplit)
    ridge_cv = RidgeCV(alphas=np.logspace(1,3.5,100), scoring='r2', cv=5)
    ridge_cv.fit(Zs_train, ys_train)
    alpha = ridge_cv.alpha_
    alphas.append(alpha)
pd.DataFrame(alphas).to_csv('datasets/tts_alphas_1k.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

plt.figure(figsize=(14,3))
plt.xlim(10, 2000)
plt.ylim(0, 100)
plt.hist(alphas, bins=25, );
#plt.xticks(np.logspace(1,3.5,100));
plt.xlabel('alpha', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('count', fontsize=15);
#plt.xscale('log')
plt.vlines(np.mean(alphas), ymin=0, ymax=100, color='green', label='mean')
plt.vlines(np.median(alphas), ymin=0, ymax=100, color='blue', label='median')
plt.vlines(80, ymin=0, ymax=100, color='red', label=' = 80')
plt.legend(fontsize=14)
```
